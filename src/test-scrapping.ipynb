```python
import requests as rs
from bs4 import BeautifulSoup as bs
```

```python
url = "https://indeks.kompas.com/terpopuler/?site=all&date=2025-05-07"
response = rs.get(url)
if response.status_code == 200:
    soup = bs(response.text, 'html.parser')
    # mengambil semua tag <h1> dari halaman urlnya
    h1_tags = soup.find_all('h1')
    for tag in h1_tags:
        print(tag.text)
else:
    print(f"Failed to retrieve page: {response.status_code}")
```

```python
# Temukan div dengan class 'articleList -list'
article_list = soup.find('div', class_='articleList -list')

# Cari semua item artikel di dalamnya (class 'articleItem')
item_populars = article_list.find_all('div', class_='articleItem') if article_list else []

data = []
for item in item_populars:
    # Mengambil link artikel
    link_tag = item.find('a', class_='article-link')
    link = link_tag['href'] if link_tag and link_tag.has_attr('href') else None
    # Mengambil gambar artikel
    image_tag = item.find('img', class_='articleImage')
    image = image_tag['src'] if image_tag and image_tag.has_attr('src') else None

    # Mengambil judul artikel
    title_tag = item.find('h2', class_='articleTitle')
    title = title_tag.get_text(strip=True) if title_tag else None

    # Mengambil kategori artikel
    category_tag = item.find('div', class_='articlePost-subtitle')
    category = category_tag.get_text(strip=True) if category_tag else None

    # Mengambil tanggal postingan artikel
    date_tag = item.find('div', class_='articlePost-date')
    date = date_tag.get_text(strip=True) if date_tag else None

    # Simpan data ke list
    data.append({
        'title': title,
        'category': category,
        'date': date,
        'link': link,
        'image': image,
    })
```

```python
import pandas as pd
# membuat variabel df untuk menampung data dari list data
df = pd.DataFrame(data)

# menampilkan data dalam bentuk tabel
print(df)
```
